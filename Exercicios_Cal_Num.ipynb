Exerc√≠cio 2

```python
n = 10
r = 0.5
k = 450

t = np.linspace(0,10,n)
p = np.linspace(100,200,n)

T,P = np.meshgrid(t,p)

f = r*P-k
ang = np.arctan(f) 
dp = np.cos(ang)
dt = np.sin(ang)

plt.quiver(T,P,dp,dt)
plt.grid('on')
```

```python
def ed(p,t):
  f = r*p-k
  return f

tt = np.linspace(0,10,10)

y0 = 100
sol = odeint(ed,y0,tt)
plt.plot(tt,sol) # plota a curva
plt.quiver(T,P,dp,dt)
plt.grid('on')
```

```python
plt.figure(figsize = (6,6),dpi=100)
plt.quiver(T,P,dp,dt,angles='xy',label='$dv/dt$')
plt.grid('on')
for v0 in range(170,180,2):
  sol = odeint(ed,v0, tt)
  plt.plot(tt,sol, label=v0)
plt.legend()
plt.xlabel('t')
plt.ylabel('p')
plt.show()
```
